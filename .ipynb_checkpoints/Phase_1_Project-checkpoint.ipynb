## Data Preparation
We, however, do not require the whole data set for our analysis as most of the information provided is not necessary for our analysis. We need to prepare the data for analysis by cleaning it.

```python
# Retain only relevant data by dropping columns that are not required in the analysis
columns = ['Event.Date', 'Country', 'Injury.Severity',
           'Aircraft.damage','Make','Model','Amateur.Built','Number.of.Engines',
            'Engine.Type','Purpose.of.flight','Weather.Condition','Broad.phase.of.flight']
df_clean = df[columns]
```

```python
# Cleaning the column names
df_clean.columns = df_clean.columns.str.replace(".","_").str.lower().str.strip()
```

```python
# Having only the years in the event_date
df_clean.loc[:,'event_date'] = df_clean.loc[:,'event_date'].str[-4:].astype('int')
```

```python
#remove null values
df_clean.dropna(inplace=True)
```

```python
#get the accidents that were not fatal
not_fatal=[f for f in df_clean['injury_severity'].unique().tolist() if f[0:5] != 'Fatal']
not_fatal
```

```python
#remove the rows with the injury_severity unavailable
df_clean = df_clean.loc[~(df_clean['injury_severity'] == 'Unavailable')]
```

```python
#reset the injury_severity to a measure by integer
df_clean.loc[:,'injury_severity'] = df_clean.loc[:,'injury_severity'].str.strip('Fatal')
df_clean.loc[:,'injury_severity'] = df_clean.loc[:,'injury_severity'].str.replace('Non-','0')
df_clean.loc[:,'injury_severity'] = df_clean.loc[:,'injury_severity'].str.replace('Inciden','0')
df_clean.loc[:,'injury_severity'] = df_clean.loc[:,'injury_severity'].str.strip('(')
df_clean.loc[:,'injury_severity'] = df_clean.loc[:,'injury_severity'].str.strip(')').astype('int')
```

```python
#check if there are any more null values in our data set
df_clean.isnull().any().sum()
```

```python
# check any duplicate entries
df_clean.duplicated().any()
```

```python
#remove duplicated rows
df_clean.drop_duplicates(inplace=True)
```

The data is cleaned and prepared to be used for analysis. We will the create a new file with our new data and use it for the analysis

```python
#create a new CSV with our new data
df_clean.to_csv('CleanedAviationData.csv')
```

## Data Analysis
Now we are going to carry out our analysis and get insights for our research.

```python
#import the data
data = pd.read_csv('CleanedAviationData.csv')
data.head()
```

```python
print(f'The cleaned data has {data.shape[0]} rows and {data.shape[1]} columns')
```

### 1. Is it safe to invest in Aviation Industry?
The aviation industry is an old industry that has evolved over the years. To know if the industry is a good investment, it is important to analyze how safer the industry is across time. This will be done by showing the number of accidents per year.

```python
#To get the number of accidents per year
number_of_accidents_per_year = data['event_date'].value_counts().tolist()
years = data['event_date'].value_counts().index.tolist()
```

```python
#import matplotlib to plot
import matplotlib.pyplot as plt
```

```python
#plot a line graph that shows the trend of accidents overtime
#I included the years from 1982 moving forward as before there were not many flights
x = years[:26]
y = number_of_accidents_per_year[:26]
fig, ax = plt.subplots(figsize = (10,8))
ax.plot(x,y)
ax.set_xlabel('Years')
ax.set_ylabel('Number of accidents')
ax.set_title('A trend of the number of accidents per year');
```

From the line graph shown above. It can be clearly seen that the number of accidents have been dropping through time, indicating that the aviation industry has been getting safer with more safety measures being adopted.

**Recommendation 1:** Seryabiz Ltd should proceed to invest in the industry as it is safer now than before and the trend indicates that it will continue to improve.

### 2. Which type of aircraft should the company invest in?
There are two types of builds that the company can invest in. It is important to know their risk.

```python
safety = data.groupby(['amateur_built'])['injury_severity'].mean().sort_values()
```

```python
x = safety.index.tolist()
y = safety.tolist()
fig,ax = plt.subplots(figsize = (10,8))
ax.bar(x,y)
ax.set_ylabel('Injury Severity')
ax.set_xlabel('Is the aircarft amateur built')
ax.set_title('The safety of amateur built and non-amateur built crafts');
```

The visual above shows that amateur built aircrafts are safer compared to aircrafts that are not amateur built.

**Recommendation 2:** The company should purpose to purchase amateur built aircrafts for the investments.

### 3. What sector should the company invest in?

```python
sector = data['purpose_of_flight'].value_counts()
```

```python
fig,ax = plt.subplots(figsize = (15,15))
fig.set_tight_layout(True)
ax.bar(sector.index.tolist()[:10],sector.tolist()[:10])
ax.set_xlabel('Purpose of flight')
ax.set_ylabel('Number of accidents and incidents')
ax.set_title('Top 10 most affected sectors in the industry');
```

**Recommendation 3:** The personal flights are the most affected sector of the aviation industry. The business sector is the advisable field as it is less affected and also has more customers compared to the other sectors.
 ### 4. What make should the company invest in?

```python
# To find the most durable makes
craft = pd.crosstab(index = data['make'],columns = data['aircraft_damage'])
craft['Total'] = craft['Destroyed']+craft['Minor']+craft['Substantial']
craft['Durability'] = craft['Destroyed']/craft['Total']
craft.sort_values('Durability',inplace = True)
```

```python
#top ten durable makes
craft.head(10)
```

**Recommendation 4:** The most durable make is Kewley and the dataframe above shows the top ten most recommended makes.
